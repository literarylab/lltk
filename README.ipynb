```python
# load corpus
import lltk
corpus = lltk.load('TxtLab')
corpus.meta
```

## Accessing metadata

```python
# show metadata
corpus.meta
```

```python
# query on metadata
corpus.meta.query('1770<=year<1790 & language=="English"')
```

```python
# examine metadata: breakdown of novel language
corpus.meta['language'].value_counts().plot.pie()
```

```python
# examine metadata: breakdown of author gender
corpus.meta['gender'].value_counts().plot.pie()
```

```python
# examine metadata: distribution of years per language
corpus.meta.groupby('language').year.plot.density()
```

## Most frequent words

```python
# Most frequent words overall
corpus.mfw_df()
```

```python
# Different settings computed on the fly
corpus.mfw_df(yearbin=10, n=10000, keep_periods=True)
```

```python
# Simple plots
import plotnine as p9
p9.options.figure_size=(8,8)
p9.ggplot(
    p9.aes(x='period',y='fpm',label='word',color='word',group='word'),
    data=corpus.mfw_df(
        texts=corpus.meta.query('language=="English"'),
        n=10,
        excl_stopwords=True,
        excl_top=100,
        yearbin=50,
        keep_periods=True
    )
) + p9.geom_line() + p9.geom_text()
```

## Document-term matrix

```python
# corpus.mfw()
```

```python
corpus.texts()[0].meta
```

```python
# Default settings
dtm = corpus.dtm(
    words=[],                          # words to use in DTM
    n=25000,                           # if not `words`, how many mfw?
    texts=None,                        # set texts to use explicitly, otherwise use all
    tf=False,                          # return term frequencies, not counts
    tfidf=False,                       # return tfidf, not counts
    meta=False,                        # include metadata (e.g. ["gender","nation"])
    # **mfw_attrs,                       # all other attributes passed to self.mfw()
)
```

```python
# Custom
tfidfs=corpus.dtm(
    texts=corpus.meta.query('language=="English"'),
    tfidf=True,
)
tfidfs
```

```python
# sort by text
tfidfs.loc['EN_1925_Woolf,Virginia_Mrs.Dalloway_Novel'].sort_values(ascending=False).head(20)
```

## Most distinctive words

```python
# Group tfidf into MDW
corpus.mdw(
    'gender',
    texts=corpus.meta.query('language=="English" & 1800<=year<1900'),
    agg='median'
)
```
